```
import pandas as pd
import numpy as np
df1 = pd.read_csv('/content/drive/My Drive/jan_analysis_2019.csv')
df1 = df1.drop('Unnamed: 0', 1)
df1.relevance=df1.relevance.astype('float64')
```

```
#list1 = ['World','U.S.','Technology','Business','Your Money']
df_world = df1[df1['section_name'].str.match('World')]
df_us = df1[df1['section_name'].str.match('U.S.')]
df_technology = df1[df1['section_name'].str.match('Technology')]
df_business = df1[df1['section_name'].str.match('Business')]
df_money = df1[df1['section_name'].str.match('Your Money')]

#df[df['model'].str.match('Mac')]
```

```
df1.head(5)
```

```
#df1['relevance'] = np.where((df1['occurence1'].str.contains('True')), 10, 0.05)
df1['occurence1'] = df1['content'].str.contains(r'Cryptocurrency|cryptocurrency|BITCOIN|bitcoin|Bitcoin')
df1['relevance1'] = df1['occurence1'].apply(lambda x:6 if x  else 0.002)
```

```
df1.head(5)
```

```
df1['occurence2'] = df1['content'].str.contains(r'economy|us dollar|money|U.S|american')
df1['relevance2'] = df1['occurence2'].apply(lambda x:0.08 if x  else 0.002)
```

```
df1.loc[31]
```

```
#1094
for i in range(1094):
  #print(i)
  if (df1['relevance1'][i] > df1['relevance2'][i]):
    df1['relevance'][i] = df1['relevance1'][i]
    #print(df1['relevance'][i])
    #print(df1['relevance1'][i])
  else:
    df1['relevance'][i] = df1['relevance2'][i]
    #print(df1['relevance2'][i])
  
```

```
df1.head(5)
```

```
df2 = df1[['web_url','content','headline','document_type','pub_date','source','section_name','author','word_count','content_no_stopwords','SENTIMENT_VALUE','SENTIMENT','relevance']]
```

```
df2.head(5)
```

```
df2['score'] = df2['SENTIMENT_VALUE']*df2['relevance']
```
