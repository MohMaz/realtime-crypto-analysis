```python
df = pd.read_csv('data_jan_2019.csv')
df1 = df[['content']]

```

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')
```

```python
df1["content_no_stopwords"]  = df1["content"] .apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
```

```python
all_reviews = df1["content_no_stopwords"]
all_sent_values = []
all_sentiments = []
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
def sentiment_value(paragraph):
    analyser = SentimentIntensityAnalyzer()
    result = analyser.polarity_scores(paragraph)
    score = result['compound']
    return round(score,1)
```

```python
sample = df1["content_no_stopwords"][200]
#print(sample)
print('Sentiment: ')
print(sentiment_value(sample))
```

```python
for i in range(0,946):
    all_sent_values.append(sentiment_value(all_reviews[i]))
print('over')
```

```python
SENTIMENT_VALUE = []
SENTIMENT = []
for i in range(0,946):
    sent = all_sent_values[i]
    if (sent<=1 and sent>=0.5):
        SENTIMENT.append('Very Positive')
        SENTIMENT_VALUE.append(sent)
    elif (sent<0.5 and sent>0):
        SENTIMENT.append('Positive')
        SENTIMENT_VALUE.append(sent)
    elif (sent==0):
        SENTIMENT.append('Neutral')
        SENTIMENT_VALUE.append(sent)
    elif (sent<0 and sent>=-0.5):
        SENTIMENT.append('Negative')
        SENTIMENT_VALUE.append(sent)
    else:
        SENTIMENT.append('Very Negative')
        SENTIMENT_VALUE.append(sent)
print('over')
```

```python
df1['SENTIMENT_VALUE'] = SENTIMENT_VALUE
df1['SENTIMENT'] = SENTIMENT
```
